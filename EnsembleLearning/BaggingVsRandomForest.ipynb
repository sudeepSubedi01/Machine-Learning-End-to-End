```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_classification
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1, random_state=42)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col15'])
df['target'] = y

print(df.shape)
print(df.head())
```

# Bagging Classifier

```python
from sklearn.ensemble import BaggingClassifier

bag = BaggingClassifier(max_features=2)
bag.fit(df.iloc[:,0:5], df.iloc[:,-1])
```

```python
bag.estimators_
```

```python
bag.estimators_[0]
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
```

Tree level sampling. Samples are decided at the root node.

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_features=2)
rf.fit(df.iloc[:,0:5], df.iloc[:,-1])
```

```python
rf.estimators_[4]
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[1])
plt.show()
```

Node level sampling.
