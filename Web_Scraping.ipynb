```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# directly request from bots is prevented in any websites
# we must make send the request as if the request is being sent by a browser
# requests.get('https://www.ambitionbox.com/list-of-companies?page=10').text
```

```python
# so we are sending headers so as to tell the server that the request is coming from a browser not a bot
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}

webpage = requests.get('https://books.toscrape.com/catalogue/page-3.html',headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'html.parser')
print(soup.prettify())
```

```python
# soup.find_all('h3') # returns a list of all the span tags
# soup.find_all('span')[0] # accessing 1st element only
# len(soup.find_all('span'))
# print(soup.find_all('h3'))
print(soup.find_all('h3')[5].text)
```

```python
for i in soup.find_all('h3'):
  # print(i) # returns 1st element of the list (1st h3 tag)
  # print(i.text)   # returns only the text inside the h3 tags
  print(i.text.strip()) #returns text inside the tag, removing all the extra spaces and newline characters
```
