```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}
webpage = requests.get('https://books.toscrape.com/catalogue/category/books/philosophy_7/index.html', headers=headers).text
# print(webpage) 
```

```python
soup = BeautifulSoup(webpage,'html.parser')
print(soup.prettify())
```

```python
soup.find_all('article', class_='product_pod')[0]
# len(soup.find_all('article', class_='product_pod'))
```

```python
# i itself is article tag; find_all method works with it
book_name = []
book_price = []
book_stock = []
for i in soup.find_all('article',class_='product_pod'):
  # print(i.text)   
  # print(i.find('h3').text.strip())
  book_name.append(i.find('h3').text.strip())
  book_price.append(i.find('p',class_='price_color').text.strip())
  book_stock.append(i.find('p',class_='availability').text.strip())
```

```python
# print(len(name))
print(book_name)
print(book_price)
print(book_stock)
```

```python
d = {'name':book_name,'price':book_price,'availability':book_stock}
df = pd.DataFrame(d)
print(df)
```
